```python
from sympy import *
from sympy.physics.secondquant import *
```

```python
a1, a2, a3 = symbols(r"\alpha_1,\alpha_2,\alpha_3")
a1p, a2p, a3p = symbols(r"\alpha_1^{'},\alpha_2^{'},\alpha_3^{'}")
a, b, g, d = symbols(r"\alpha,\beta,\gamma,\delta")
```

```python
expr = F(a3)*F(a2)*F(a1)*Fd(a)*Fd(b)*F(d)*F(g)*Fd(a1p)*Fd(a2p)*Fd(a3p)
```

```python
expr
```

```python
w = wicks(expr, keep_only_fully_contracted=True)
w
```

```python
ans = simplify(w)
ans
```
