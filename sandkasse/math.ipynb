```python
import sympy as sp
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
sp.init_printing()
```

```python
g, d = sp.symbols("g,d")
```

```python
H = sp.Matrix([
    [2*d-g, -g],
    [-g, 4*d-g]
])
```

```python
vals = H.eigenvals()
vals
```

```python
for elm in H.eigenvects():
    val, _, vec = elm
    vec, = vec
    display(val)
    display(sp.simplify(vec))
    print("------")
```

```python
l = sp.symbols(r"\lambda")
det = -(6*d-2*g)*l + l**2 + (2*d-g)*(4*d-g) -g**2
```

```python
res1 = det.subs(l, val)
res1 = sp.simplify(res1)
res1
```
